```python
x = np.arange(-10, 10, 0.01)
y = x**2 + np.random.randn(len(x))*x
plt.scatter(x, y, s=0.3)
p = np.polyfit(x, y, deg=2)
p_f = np.poly1d(p)
plt.plot(x,p_f(x))
plt.show()
```

```python
from math import cos, pi

a = int(input('a:'))
b = float(input('b:'))
n = int(input('n:'))

x_data = np.arange(-2, 2.01, 0.01)
y_data = np.array([])
for i in x_data:
    y=0
    for j in range(n):
        y += b**j*cos(a**j*pi*i)
    y_data = np.append(y_data, y)

plt.figure(figsize = (16, 9))
plt.plot(x_data,y_data)
```
