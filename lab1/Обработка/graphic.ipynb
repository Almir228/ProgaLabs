```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

Бинарный поиск

```python
data1 = pd.read_csv("data/data1.csv", delimiter=";", names=['N', 'Bin_Search'])

# Bin_search = Bin_search[Bin_search["Bin_search"] < 1e-5]
data1
```

Апроксимация логрифмом

```python
def logarifm(x, a, b, c):
    return a*np.log(b+x)+c

popt, _ = curve_fit(logarifm, data1["N"].tolist(), data1["Bin_Search"].tolist())

a, b, c = popt

y = logarifm(data1['N'], a, b, c)
```

```python
fig, ax = plt.subplots()
ax.scatter(data1["N"], data1["Bin_Search"])
ax.plot(data1['N'], y)

plt.title("Бинарный поиск")
ax.set_xlabel("Число элементов в массиве")
ax.set_ylabel("Время обработки")
plt.savefig("data/Bin_Search.pdf")
```

Линейный поиск

```python

data2 = pd.read_csv("data/data2.csv", delimiter=";", names=['N', 'Search'])

# Bin_search = Bin_search[Bin_search["Bin_search"] < 1e-5]
data2
```

Линейная апроксимация

```python
def line(x, a, b):
    return a*x+b

popt, _ = curve_fit(line, data2["N"].tolist(), data2["Search"].tolist())

a, b = popt

y = line(data2['N'], a, b)
```

```python
a = plt.subplot()
a.scatter(data2["N"], data2["Search"], label='high income low saving', color='b')
a.plot(data2['N'], y)
plt.title("Линейный поиск")
a.set_xlabel("Число элементов в массиве")
a.set_ylabel("Время обработки")
plt.savefig("data/Search.pdf")
```
